7. Is it true that people who earn more than 50K have at least high school education? 

```python
education_order = [
    "Preschool", "1st-4th", "5th-6th", "7th-8th",
    "9th", "10th", "11th", "12th",
    "HS-grad", "Some-college", "Assoc-voc", "Assoc-acdm",
    "Bachelors", "Masters", "Doctorate", "Prof-school"
]

education_rank = {edu: i for i, edu in enumerate(education_order)}
data["education_rank"] = data["education"].map(education_rank)

high_earners = data[data["salary"].str.contains(">50K")]

below_hs = high_earners[high_earners["education_rank"] < education_rank["HS-grad"]]

num_below_hs = below_hs.shape[0]
num_high_earners = high_earners.shape[0]
percentage = (num_below_hs / num_high_earners) * 100
print(f"Total people earning >50K: {num_high_earners}")
print(f"People earning >50K with education below HS-grad: {num_below_hs}")
print(f"Percentage of exceptions: {percentage:.2f}%")

if num_below_hs == 0:
    print("\n✅ All people earning >50K have at least high school education.")
else:
    display(below_hs[["education", "salary"]].head())
```

8. Display age statistics for each race (race feature) and each gender (sex feature). Use groupby() and describe(). Find the maximum age of men of Amer-Indian-Eskimo race.

```python
age_stat = data.groupby(['race', 'sex'])['age'].agg(['mean', 'min', 'max'])

AIE_max = data[(data['race'] == 'Amer-Indian-Eskimo') & (data['sex'] == 'Male')]['age'].max()

display(age_stat)
print('maximum age of men of Amer-Indian-Eskimo race')
print(AIE_max)
```

Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (marital-status feature)? Consider as married those who have a marital-status starting with Married (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.

```python
married_man = data[
    (data['marital-status'].str.startswith('Married')) &
    (data['sex']== 'Male')
    ]
ststs = ((married_man['salary'] == '>50K').sum())
print('>50K' , ststs)
print('<=50K' , married_man.shape[0] - ststs)
```

How many people work maximum number of hours, and what is the percentage of those who earn a lot (>50K) among them?

```python

max_hours = data['hours-per-week'].max()
max_workers = data[data['hours-per-week'] == max_hours]
num_max_workers = len(max_workers)
num_rich_max_workers = (max_workers['salary'] == '>50K').sum()
percent_rich = (num_rich_max_workers / num_max_workers) * 100
print(f"Number of people who work max hours: {num_max_workers}")
print(f"Percentage earning >50K among them: {percent_rich:.2f}%")
```

Count the average time of work (hours-per-week) for those who earn a little and a lot (salary) for each country (native-country). What will these be for Japan?

```python
country_avg_hours = data.groupby(['native-country', 'salary'])['hours-per-week'].mean().reset_index()

# Display results for all countries
print("Average hours per week by country and salary group:")
display(country_avg_hours)

# Show specifically Japan's results
display("\nDetailed results for Japan:")
display(country_avg_hours[country_avg_hours['native-country'] == 'Japan'])
```
